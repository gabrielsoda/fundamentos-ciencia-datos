Escribir una función para realizar una interpolación lineal por tramos para los datos de la tabla de abajo. La función recibe como input el valor de x y como output el valor de y correspondiente:

|     x    |    y     |
|----------|----------|
|    1     |    2     |
|    2     |    3     |
|    3     |    5     |
|    10    |    6     |


```python
import pandas as pd

data = pd.DataFrame({
    'x': [1, 2, 3, 10],
    'y': [2, 3, 5, 6]
})
def interp_lineal(x_val, df):
 # df.sort_values(by='x', inplace=True)
  if x_val < df['x'].min() or x_val > df['x'].max():
    return "X está fuera del rango posible de interpolación"

  for i in range(len(df) - 1):
    x1, y1 = df.iloc[i]
    x2, y2 = df.iloc[i + 1]

    if x1 <= x_val <= x2:
      y_val = y1 + (y2 - y1) * (x_val - x1) / (x2 - x1)
      return y_val
  return "Algo salió mal"
```

```python
"""
x	y
1	2
2	3
3	5
10	6
"""
print("Debería estar entre 3 y 5, y da: ", interp_lineal(8, data))  # Debería estar entre 3 y 5
print("Entre 5 y 6, y da: ",interp_lineal(5, data))    # Entre 5 y 6
print("Debe dar 6, y da: ",interp_lineal(10, data))   # Debe dar 6
```
